##Convolution Operation
A convolution a simple mathematical operation on an input image, using a general purpose filter. In image processing, the filter is Two-dimensional and is called a **kernel**.

##Kernel
Differently sized kernels containing different patterns of numbers produce different results under convolution. The arbitrary size for a Kernel is 3x3. Kernel patterns depend upon the feature to be extracted.

The convolution is performed by sliding the kernel over the image, generally starting at the top left corner, so as to move the kernel through all the positions where the kernel fits entirely within the boundaries of the image.

#Convolution Parameters
Apart from Kernel patterns, we decide other paramers for convolutions. 
Let's take a look at Convolution parameters before we see them in action.

##Kernel Size

Kernel is used in a Convolutional Neural Network to extract useful information. A standard kernel is of size 3x3x1 for grayscale images and 3x3x3 for colored images.

##Stride

Stride is the number of units shifted by the kernel per slide. 

stride.gif

##Normalization

When the spread of inputs are different.
Example: Batch Normalization is performed to reduce the variation between weights.


##Padding

Padding adds rows and columns to the input image to ease the extraction process from the boundaries of the image.

padding.png


##Optimizer

Optimizers are algorithms or methods used to change the attributes of your neural network such as weights and learning rate in order to reduce the losses. Example: Adam, SGD, etc.


##Activation Function

Activation functions are mathematical equations that determine the output of a neural network. 
Example: Sigmoid, ReLu, leaky ReLu, etc.

activation.png

##Pooling Layer

Pooling progressively reduces the spatial size of the representation by decreasing the amount of parameters and computation in the network. 
Example: MaxPooling, AveragePooling and SumPooling.

#A brief intro to 3D convolutions

A 3D Convolution is a type of convolution where the kernel slides in 3 dimensions as opposed to 2 dimensions with 2D convolutions. One example use case is medical imaging where a model is constructed using 3D image slices. Additionally, 3D convolutions can be used for video based data and other 3D datasets.
